# Optimal masking photometry test

Here we test whether optimal masking photometry might work better than simple aperture photometry.

**Results:** Nope, aperture photometry seems to be the way to go. The contribution of the sky is a minor factor.

## Create the aperture mask

```python
def create_mask(flux, sky, rdns, f=1.):
    signal = (flux-sky).ravel()
    sid  = argsort(mf(flux-sky,2).ravel())[::-1]
    
    ## Cumulative signal
    cums = (signal[sid]).cumsum()
    
    ## Cumulative noise: sqrt(ns_sig**2 + ns_sky**2 + ns_rd**2)
    cumn = sqrt((flux.ravel() + rdns**2)[sid].cumsum()) 
    
    mask = zeros(signal.size, np.bool)
    mask[sid[:floor(f*argmax(cums/cumn))]] = 1
    return mask.reshape(flux.shape)
```

```python
mask1 = dstack([create_mask(ccd1[:,:,i], sky2d1[:,:,i], rdns, 3) for i in range(ccd1.shape[2])])
```

```python
mask2 = dstack([create_mask(ccd2[:,:,i], sky2d2[:,:,i], rdns, 3) for i in range(ccd2.shape[2])])
```
