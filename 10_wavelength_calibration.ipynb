## The IRAF solution

Calculated using the ``ecidentify`` IRAF task with Chebychev polynomials.

```python
def cheby(nmax,x,xmin,xmax):
    p = zeros((max(2,nmax),x.size))
    p[0,:] = 1.0
    p[1,:] = (2*x - (xmax + xmin)) / (xmax - xmin)
    for i in range(2,nmax):
        p[i,:] = 2*p[1,:]*p[i-1,:] - p[i-2,:]
    return p

coeffs1 = array([[11393.66802,3874.86981755,165.745593446,-18.49966226,0.786503],
                 [3791.828657,1292.57377688,55.3944656656,-6.212540604,0.257664]]).T
coeffs2 = array([[11390.5624,3844.28691,162.70662,-18.69386,0.685879],
                 [3817.60567,1267.50489,53.142094,-6.027987,0.290123]]).T
```

```python
def map_to_wl(coeffs, oo=0):
    xmin,xmax = 1,1990
    omin,omax = 1,2
    x = arange(xmin,xmax+1, dtype=np.float64)
    o = arange(omin,omax+1, dtype=np.float64)
    
    Px = cheby(5, x, xmin,xmax)
    Po = cheby(2, o, omin,omax)

    t = zeros_like(x)
    for m in range(5):
        for n in range(2):
            t += coeffs[m,n]*Px[m,:]*Po[n,oo]
    return t*1e-1
```

```python
wl1 = map_to_wl(coeffs1)+2.5
wl2 = map_to_wl(coeffs2)+6.0
```

```python
fig,ax = subplots(1,1,figsize=(13,6))
plot(wl1, arc1[31:-30,sl1,1].mean(1))
[axvline(l, c='r') for l in ne_lines]
#xlim(570,800)
fig.tight_layout()
```

```python
fig,ax = subplots(1,1,figsize=(13,6))
plot(wl2, arc2[31:-30,sl2,1].mean(1))
[axvline(l, c='r') for l in ne_lines]
#xlim(570,800)
fig.tight_layout()
```
