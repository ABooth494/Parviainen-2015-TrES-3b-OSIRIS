```python
from scipy.optimize import fmin
from pytransit import MandelAgol
from pytransit.orbits_f import orbits as of
from exotk.utils.orbits import i_from_baew
from exotk.utils.limb_darkening import quadratic_law

tt = load('data/tres_3_limb_darkening_c.npz')
z2 = sqrt(1-tt['mu']**2)
i2 = tt['im']

k1,k2 = 0.185, 0.175
l1,l2 = [0.58,0.16], [0.48,0.19]

zld = linspace(0,1)
mu  = sqrt(1-zld**2)

tm = MandelAgol()
p,a,b = 1.3,6.03, 0.84
i = i_from_baew(b,a,0,0)
t = linspace(0.96,1.04)
z = of.z_circular(t, 1, p, a, i, 0)

fo1 = tm(z,k1,l1)
fo2 = tm(z,k2,l2)
```

We estimate $k=0.185$ with limb darkening coefficients $u=0.58$ and $v=0.16$ for the passband centred around 542.5 nm, and $k=0.175$ with $u=0.48$ and $v=0.19$ for the passband centred around 642.5 nm. Can the difference in radius radii be explained by systematic errors in limb darkening?

```python
plot(t,tm(z, k1, l1), 'b')
plot(t,tm(z, k2, l2), 'r')
xlim(t[[0,-1]])
ylim(0.97,1.001);
```

Can the large blue radius ratio be explained with a smaller radius ratio and different limb darkening?

```python
res1 = fmin(lambda ld:((fo1-tm(z, k2, ld))**2).sum(), l1)
res1
```

```python
fig,ax = subplots(1,2,figsize=(14,4))
ax[0].plot(t,fo1, 'k')
ax[0].plot(t,tm(z, k2, res1), 'g')
ax[1].plot(z2,i2[:,0], 'k--', lw=4, alpha=0.5)
ax[1].plot(zld, quadratic_law(mu, l1), 'k')
ax[1].plot(zld, quadratic_law(mu, res1))
setp(ax[0], xlim=t[[0,-1]], ylim=(0.97,1.001))
fig.tight_layout()
```

Can the estimated redder smaller radius ratio be explained with a larger true $k$ and different limb darkening?

```python
res2 = fmin(lambda ld:((fo2-tm(z, k1, ld))**2).sum(), l2)
res2
```

```python
fig,ax = subplots(1,2,figsize=(14,4))
ax[0].plot(t,fo2, 'k')
ax[0].plot(t,tm(z, k1, res2), 'g')
ax[1].plot(z2,i2[:,4], 'k--', lw=4, alpha=0.5)
ax[1].plot(zld, quadratic_law(mu, l2), 'k')
ax[1].plot(zld, quadratic_law(mu, res2))
setp(ax[0], xlim=t[[0,-1]], ylim=(0.97,1.001))
fig.tight_layout()
```

```python
def estimate_ld(k0):
    def minfun(pv):
        return ((fo1-tm(z, k0, pv[:2]))**2).sum() + ((fo2-tm(z, k0, pv[2:]))**2).sum()
    return fmin(minfun, [0.58,0.16, 0.48,0.19])
```

```python
k0 = 0.18
res3 = estimate_ld(k0)
```

```python
k0 = 0.18
res3 = estimate_ld(k0)

fig,ax = subplots(1,2,figsize=(14,4))

ax[0].plot(t,fo1, 'g')
ax[0].plot(t,tm(z, k0, res3[:2]), 'g')
ax[0].plot(t,fo2, 'g')
ax[0].plot(t,tm(z, k0, res3[2:]), 'g')

ax[1].plot(zld, quadratic_law(mu, l1), 'b-')
ax[1].plot(zld, quadratic_law(mu, res3[:2]), 'b--')

ax[1].plot(zld, quadratic_law(mu, l2), 'r-')
ax[1].plot(zld, quadratic_law(mu, res3[2:]), 'r--')

setp(ax[0], xlim=t[[0,-1]], ylim=(0.97,1.001))
fig.tight_layout()
```
