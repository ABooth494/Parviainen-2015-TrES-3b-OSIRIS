## Fit

```python
class LPFunction(object):
    """A simple log posterior function class.
    """
    def __init__(self, time, flux_w, fluxes, airmass, nthreads=2):
        self.tm = MA(lerp=True, klims=(0.14,0.20), nthr=nthreads) 
        self.nthr = nthreads

        self.time     = array(time)
        self.flux_w   = array(flux_w)
        self.fluxes   = array(fluxes)
        self.airmass  = array(airmass)
        self.fratio   = self.fluxes/self.flux_w
        self.npt      = self.flux_w.size
        self.npb      = self.fluxes.shape[0]
        self._wrk_lc  = zeros_like(time)
        self._wrk_ld  = zeros(2)
        
        self.hpv = array([ 0.00114895,  0.02112864,  0.00181875])
        
        self.gp = GP(self.hpv[0]**2*ExpKernel(self.hpv[1]))
        self.gp.compute(self.airmass, self.hpv[2])
        
        self.oot_mask = ones(self.npt, np.bool)
        self.oot_mask[45:185] = 0
        
        self.baselines = zeros([self.npt, self.npb+1])
        self.models    = zeros([self.npt, self.npb+1])
        
        self.priors = [NP(56846.11,  0.005,  'tc'), ##  0  - Transit centre
                       NP(1.30619,  1.1e-5, 'p'),   ##  1  - Period
                       UP( 0.15**2,0.19**2,  'k2'), ##  2  - area ratio
                       NP(  1.646, 0.05,  'rho', lims=[1,3]), ## 3  - Stellar density
                       #NP( 6.030,  0.176,   'a'), ##  3  - Scaled semi-major axis
                       #NP( 0.827,  0.010,   'b'), ##  4  - Impact parameter
                       UP( 0.000,  0.995,   'b'), ##  4  - Impact parameter
                       UP(  5e-4,   6e-3,   'e'), ##  5  - White noise std
                       NP(   1.0,   0.01,   'c'), ##  6  - Baseline constant
                       NP(  0.40,   0.01,   'v'), ##  7  - limb darkening u
                       NP(  0.20,   0.02,   'w'), ##  8  - limb darkening v
                       UP(-0.010,   0.01,   'ex')] ##  9  - extinction coefficient
        
        for i in range(self.npb):
            self.priors.extend([
                       UP( 0.900,  1.100,  'kf'), ## 10 + 5*ipb  - Narrow band k2 factor
                       NP(  1.00,   0.01, 'frb'), ## 11 + 5*ipb  - Narrow band baseline
                       UP(  1e-3,   3e-3, 'fre'), ## 12 + 5*ipb  - Flux ratio scatter
                       NP(  1.00,   0.10, 'fff'), ## 13 + 5*ipb  - 
                       NP( -0.01,   0.01, 'ext')] ## 14 + 5*ipb  -
            )
        self.ps = PriorSet(self.priors)
        self._wrk_pv  = zeros(self.ps.ndim)
        
        mo = self.oot_mask
        
        def minfun(pv, i):
            if i==-1:
                self._wrk_pv[6] = pv[0]
                self._wrk_pv[9] = pv[1]
                return sum((self.flux_w[mo]/self.compute_baseline(self._wrk_pv)[mo,0]-1)**2)
            else:
                self._wrk_pv[11+i*5] = pv[0]
                self._wrk_pv[14+i*5] = pv[1]
                return sum((self.fluxes[i,mo]/self.compute_baseline(self._wrk_pv)[mo,i+1]-1)**2)

        self.pbl = [fmin(minfun, [1,0], args=(ipb,), disp=False) for ipb in arange(-1,self.npb)]

        
    def compute_baseline(self, pv, time=None):
        self.baselines[:,0] = pv[6]*exp(pv[9]*self.airmass)
        for ipb in range(self.npb):
            self.baselines[:,ipb+1] = pv[11+5*ipb]*exp(pv[14+ipb*5]*self.airmass)
        return self.baselines
    
    
    def compute_transit(self, pv, time=None):
        """Transit model"""
        _a = as_from_rhop(pv[3], pv[1])
        _i  = mt.acos(pv[4]/_a)
        #_i  = mt.acos(pv[4]/pv[3])
        _k  = mt.sqrt(pv[2])
        
        time = self.time if time is None else time
        z = of.z_circular(time, pv[0], pv[1], _a, _i, nthreads=self.nthr)
        
        self.models[:,0] = self.tm(z, _k, pv[7:9])
        for ipb in range(self.npb):
            self.models[:,ipb+1] = self.tm(z, pv[10+5*ipb]*_k, pv[7:9])
        return self.models

    
    def compute_lc_model(self, pv, time=None):
        """Combined baseline and transit model"""
        return self.compute_baseline(pv, time) * self.compute_transit(pv, time)

    def compute_flux_ratios(self, pv, ipb, flux_m):
        b = pv[13+ipb*5]
        fro = self.fluxes[ipb] / (1.+b*(self.flux_w-1.))
        frm = flux_m[:,1+ipb]  / (1.+b*(flux_m[:,0]-1.))
        return fro, frm
        
    def __call__(self, pv):
        """Log posterior density"""
        if np.any(pv < self.ps.pmins) or np.any(pv>self.ps.pmaxs):
            return -inf
        flux_m = self.compute_lc_model(pv)
        ll_w   = ll_normal_es(self.flux_w, flux_m[:,0], pv[5])
        ll_n   = sum([ll_normal_es(*self.compute_flux_ratios(pv, ipb, flux_m), 
                                   e=pv[12+ipb*5]) for ipb in range(self.npb)])
        
        
        return self.ps.c_log_prior(pv) + ll_w + ll_n
```

```python
lpf = LPFunction(mjd,mean(lcr_n,0),lcr_n,df.airmass)
```

```python
de = DiffEvol(lpf, lpf.ps.bounds, 200, maximize=True)
sampler = EnsembleSampler(200, lpf.ps.ndim, lpf)
```

```python
de._population[:,6] = normal(lpf.pbl[0][0], 0.001, size=200)
de._population[:,9] = normal(lpf.pbl[0][1], 0.001, size=200)
for ipb in range(lpf.npb):
    de._population[:,11+5*ipb] = normal(lpf.pbl[ipb+1][0], 0.001, size=200)
    de._population[:,14+5*ipb] = normal(lpf.pbl[ipb+1][1], 0.001, size=200)
```
