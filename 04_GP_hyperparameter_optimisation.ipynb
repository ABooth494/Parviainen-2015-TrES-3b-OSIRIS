```python
%pylab inline
import seaborn as sb
from mpl_toolkits.axes_grid1 import make_axes_locatable

from core import *

from fit_color import GLPF_WN as LPFW

from scipy.optimize import fmin
from pyde.de import DiffEvol 

from george import GP
from george.kernels import ExpKernel
```

## Load the results

```python
lpf = LPFW() 
ch = load('results/TrES_3b_color_wn_mc.npz')['chains']
fc = ch.reshape((-1,lpf.ps.ndim))
```

```python
pv = median(fc, 0)
fo = lpf.lpf.normalize_flux(pv)
fm = lpf.lpf.compute_lc_model(pv)
residuals = fo-fm
scatter = residuals.std(0)
```

```python
gp = GP(ExpKernel(1))
time = lpf.lpf.time.ravel()
```

```python
def gp_mean(pv, ipb):
    gp.kernel = pv[0]**2*ExpKernel(pv[1])
    gp.compute(time, pv[2+ipb])
    return gp.predict(residuals[:,ipb], time, mean_only=True)

def minfun(pv):
    if any(pv<0): return np.inf
    ll = 0
    gp.kernel = pv[0]**2*ExpKernel(pv[1])
    for ipb in range(lpfw.npb):
        gp.compute(time, pv[2+ipb])
        ll += gp.lnlikelihood(residuals[:,ipb])
    return -ll
```

```python
bounds = zeros([18,2])
bounds[:,0] = concatenate([[0.01, 0.0003],0.2*scatter])
bounds[:,1] = concatenate([[0.10, 0.0015],1.2*scatter])
```

```python
de = DiffEvol(minfun, bounds, 100, C=0.85)
```

```python
de.optimize(30)
```

savez('results/gp_color_250_hps', hps=de.minimum_location, population=de.population, fitness=de._fitness)
