```python
class LPFunction(object):
    """A simple log posterior function class.
    """
    def __init__(self, time, fluxes, airmass, nthreads=2):
        self.tm = MA(lerp=True, klims=(0.15,0.22), nthr=nthreads, nk=128) 
        self.nthr = nthreads

        self.time     = array(time)
        self.fluxes   = array(fluxes)
        self.airmass  = tile(airmass, [fluxes.shape[0],1])
        self.npt      = self.fluxes.shape[1]
        self.npb      = self.fluxes.shape[0]
        self._wrk_lc  = zeros_like(time)
        
        self.oot_mask = ones(self.npt, np.bool)
        self.oot_mask[45:185] = 0
        
        self.baselines = zeros_like(self.fluxes)
        self.models    = zeros_like(self.fluxes)
        
        self.priors = [NP(56846.11,  0.005,  'tc'), ##  0  - Transit centre
                       NP(1.30619,  1.1e-5, 'p'),   ##  1  - Period
                       UP( 0.16**2,0.20**2,  'k2'), ##  2  - average area ratio
                       NP( 1.646,  0.050,   'rho', lims=(1,3)), ##  3  - stellar density
                       NP( 0.840,  0.010,   'b', lims=(0,0.99))    ##  4  - impact parameter
                       ] 
        
        self.start_kr = len(self.priors)
        self.kr_sl = s_[self.start_kr:self.start_kr+self.npb-1]
        for i in range(self.npb-1):
            self.priors.append(UP( 0.500,  1.500,  'kf')) ## 5 + 2*ipb  - Narrow band k2 factor

        self.start_ff = len(self.priors)
        self.ff_sl = s_[self.start_ff:self.start_ff+self.npb-1]
        for i in range(self.npb-1):
            self.priors.append(NP(  1.00,   0.10, 'fff')) ## 6 + 2*ipb  - 
            
        self.start_e = len(self.priors)
        for i in range(self.npb):
            self.priors.append(UP(  1e-3,   5e-3, 'fre')) ## 12 + 5*ipb  - Flux ratio scatter
            
        self.start_blc = len(self.priors)
        self.blc_sl = s_[self.start_blc:self.start_blc+self.npb]
        for i in range(self.npb):
            self.priors.append(NP( 1.0,   0.01,   'blc'))  ##  7  - Baseline constant    

        self.start_ble = len(self.priors)
        self.ble_sl = s_[self.start_ble:self.start_ble+self.npb]
        for i in range(self.npb):
            self.priors.append(UP(-0.010,   0.01,   'ex'))  ##  7  - extinction coefficient

        self.start_ldc = len(self.priors) 
        self.ld_sl = s_[self.start_ldc:self.start_ldc+2*self.npb]
        for i in range(self.npb):
            self.priors.extend([
                       UP( -10.0,   10.0,   'v'), ##  7  - limb darkening u
                       UP( -10.0,   10.0,   'w')] ##  8  - limb darkening v
            )

        self.ps = PriorSet(self.priors)

                
        ## Load the stellar brightness profiles
        ##        
        ldc = load('data/tres_3_limb_darkening_c.npz')
        ldw = load('data/tres_3_limb_darkening_w.npz')
        
        self.im = vstack([ldw['im'][:-1],ldc['im'][:-1,:].T])
        self.ie = vstack([ldw['ie'][:-1],ldc['ie'][:-1,:].T])
        self.iv = self.ie**2
        self.mu = ldw['mu'][:-1]

        wrk_pv = zeros(self.ps.ndim)
        mo = self.oot_mask
        def minfun(pv, i):
            wrk_pv[self.start_blc+i] = pv[0]
            wrk_pv[self.start_ble+i] = pv[1]
            return ((self.fluxes[i,mo]-self.compute_baseline(wrk_pv)[i,mo])**2).sum()

        self.ld0 = [fmin(lambda pv: ((im - quadratic_law(self.mu, pv))**2).sum(), [0.5, 0.2], 
                         disp=False) for im in self.im]
        self.pbl = [fmin(minfun, [1,0], args=(ipb,), disp=False) for ipb in arange(self.npb)]

        
    def compute_baseline(self, pv, time=None):
        self.baselines = pv[self.blc_sl,newaxis]*exp(pv[self.ble_sl,newaxis]*self.airmass)
        return self.baselines
    
    
    def ld_log_likelihood(self,pv):
        i_model = quadratic_law(self.mu, pv[self.ld_sl].reshape((self.npb,2)))
        chi_sqr = ((self.im - i_model)**2/self.iv).sum()
        log_l   = -0.5*self.im.size*np.log(2*pi) -np.log(self.ie).sum() - 0.5*chi_sqr
        return log_l

    
    def compute_transit(self, pv, time=None):
        """Transit model"""
        _a  = as_from_rhop(pv[3], pv[1])
        _i  = mt.acos(pv[4]/_a)
        _k  = mt.sqrt(pv[2])
        
        time = self.time if time is None else time
        z = of.z_circular(time, pv[0], pv[1], _a, _i, nthreads=self.nthr)
        
        self.models[:,:] = self.tm(z, _k, pv[self.ld_sl]).T
        self.models[1:,:] = 1. + (self.models[1:,:]-1.)*pv[self.kr_sl,newaxis]**2
        return self.models

    
    def compute_lc_model(self, pv, time=None):
        """Combined baseline and transit model"""
        return self.compute_baseline(pv, time) * self.compute_transit(pv, time)

    
    def compute_flux_ratios(self, pv, ipb, flux_m):
        b = pv[self.start_ff+ipb-1]
        fro = self.fluxes[ipb] / (1.+b*(self.fluxes[0]-1.))
        frm = flux_m[ipb]      / (1.+b*(flux_m[0]-1.))
        return fro, frm
        
        
    def __call__(self, pv):
        """Log posterior density"""
        if np.any(pv < self.ps.pmins) or np.any(pv>self.ps.pmaxs):
            return -inf
        flux_m = self.compute_lc_model(pv)
        ll_w   = ll_normal_es(self.fluxes[0], flux_m[0], pv[self.start_e])
        ll_n   = sum([ll_normal_es(*self.compute_flux_ratios(pv, ipb, flux_m), 
                                   e=pv[self.start_e+ipb]) for ipb in range(1,self.npb)])
        ll_ld  = self.ld_log_likelihood(pv)
        
        
        return self.ps.c_log_prior(pv) + ll_w + ll_n + ll_ld
```
