```python
def predictive_model(lpf, fc, nsamples=1000):
    npt = lpf.lpf.npt
    d_baseline   = array([lpf.lpf.compute_baseline(pv) for pv in permutation(fc)[:nsamples]]).reshape((nsamples,npt))
    d_transit    = array([lpf.lpf.compute_lc_model(pv) for pv in permutation(fc)[:nsamples]]).reshape((nsamples,npt))
    est_baseline = array(percentile(d_baseline, [50,1,99], axis=0))
    est_model    = array(percentile(d_baseline*d_transit, [50,1,99], axis=0))
    
    return est_baseline, est_model
```

### Load the results

```python
lpfw = LPFW()                                        ## White noise log posterior function
lpfr = LPFR()                                        ## Red noise log posterior function
df  = import_white_lc()                              ## Pandas dataframe with all the data

chw  = load('results/TrES_3b_white_wn_mc.npz')['chains'] ## White noise MCMC chains
chr  = load('results/TrES_3b_white_gp_mc.npz')['chains'] ## Red noise MCMC chains
chwb = load('results/TrES_3b_white_wn_ib_mc.npz')['chains'] ## White noise MCMC chains
chrb = load('results/TrES_3b_white_gp_ib_mc.npz')['chains'] ## Red noise MCMC chains

fcw  = chw.reshape((-1,chw.shape[-1]))                           ## Flattened white noise chains
fcr  = chr.reshape((-1,chr.shape[-1]))                           ## flattened red noise chains
fcwb = chwb.reshape((-1,chwb.shape[-1]))                         ## Flattened white noise chains
fcrb = chrb.reshape((-1,chrb.shape[-1]))                         ## Flattened red noise chains
```

### Create the dataframes

```python
dfw, dfwb, dfr, dfrb = [pd.DataFrame(fc, columns=lpf.ps.names) 
                  for fc,lpf in zip([fcw,fcwb,fcr,fcrb],[lpfw,lpfw,lpfr,lpfr])]

dfw['k'] = sqrt(dfw.k2_0)
dfw.drop(['p','k2_0'], 1, inplace=True)

dfwb['k'] = sqrt(dfwb.k2_0)
dfwb.drop(['p','k2_0'], 1, inplace=True)

dfr['k'] = sqrt(dfr.k2_0)
dfr['gp_scale_length'] = 1./exp(dfr.gp_log_inv_length)
dfr.drop(['p','k2_0'], 1, inplace=True)

dfrb['k'] = sqrt(dfrb.k2_0)
dfrb['gp_scale_length'] = 1./exp(dfrb.gp_log_inv_length)
dfrb.drop(['p','k2_0'], 1, inplace=True)
```

### Calculate the parameter estimates

```python
from collections import OrderedDict
```

```python
params = OrderedDict(k=('Radius ratio','$\srad$'),
                     tc=('Transit center','d'), 
                     rho=('Stellar density', '$\gcm$'),
                     b=('Impact parameter','$\srad$'), 
                     ldc_0_0=('Limb darkening u',''), 
                     ldc_0_1=('Limb darkening v',''),
                     el_0=('White noise $\sigma$','ppm'),
                     gp_std=('GP $\sigma$','ppm'), 
                     gp_scale_length=('GP scale','d'))
```

```python
print dfr.gp_scale_length.mean()
print dfr.gp_std.mean()
print dfr.el_0.mean()
```

```python
for c,cname in params.items():
    ests = ["{:20s} & {:9s}".format(cname[0], '['+cname[1]+']')]
    if c in dfw.columns:
        for d in [dfw,dfwb]:
            est = array(percentile(d[c], [50, 16, 84])) * (1e6 if c in ['el_0','gp_std'] else 1)
            med, err = est[0], mean(abs(est[1:]-est[0]))
            ests.append("{:6.5f} $\pm$ {:6.5f}".format(med,err))
    else:
        ests.extend(['',''])
           
    for d in [dfr,dfrb]:
            est = array(percentile(d[c], [50, 16, 84])) * (1e6 if c in ['el_0','gp_std'] else 1)
            med, err = est[0], mean(abs(est[1:]-est[0]))
            ests.append("{:6.5f} $\pm$ {:6.5f}".format(med,err))
            
    print ' & '.join(ests)
```

```python
def hexbin2(fig, ax, a1, a2, b1, b2, alim, blim, gs):
    h1 = ax.hexbin(a1, b1, extent=(alim[0],alim[1],blim[0],blim[1]), 
                   gridsize=gs, cmap=plt.cm.Blues)
    h2 = ax.hexbin(a2, b2, extent=(alim[0],alim[1],blim[0],blim[1]), 
                   gridsize=gs, cmap=plt.cm.Greens)
    fig.canvas.draw()
    fc = h1.get_facecolors() + h2.get_facecolors()
    fc *= 0.5
    fc[fc[:,:3].sum(1)>2.9,:] = array([1,1,1,1])
    h2.set_facecolors(fc)
    fig.canvas.draw()
    ax.set_xlim(*klim)
    ax.set_ylim(*blim)
```

```python
lpfr.gp.kernel = fcr[0,10]**2*ExpKernel(1./10**fcr[0,11])
lpfr.gp.compute(lpfr.lpf.time, fcr[0,5])

obs,mod,bsl,gpm = [],[],[],[]
for pv in permutation(fcr)[:100]:
    obs.append(lpfr.lpf.normalize_flux(pv).ravel())
    bsl.append(lpfr.lpf.compute_baseline(pv).ravel())
    mod.append(lpfr.lpf.compute_lc_model(pv).ravel())
    gpm.append(lpfr.gp.predict(obs[-1]-mod[-1], time.ravel(), mean_only=True))
obs,mod,bsl,gpm = map(array, [obs,mod,bsl,gpm])
```
