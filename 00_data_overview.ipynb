```python
import math as m
import pandas as pd
import pyfits as pf
from plots import AAOCW, AAPGW
import seaborn as sb

from core import *

rc('font', size=6)
rc('axes', labelsize=8)
rc('ytick', labelsize=8)
rc('xtick', labelsize=8)
rc('axes', titlesize=10)

datadir = '../../../data/GTC9-14A/OB0008/'
```

## Spectra

```python
b=10
psl1,psl2 = s_[540:560],s_[405:425] 
wsl1,wsl2 = s_[psl1.start-b:psl1.stop+b], s_[psl2.start-b:psl2.stop+b]
ssl1,ssl2 = s_[wsl1.start-b:wsl1.start],  s_[wsl2.start-b:wsl2.start]

wl1,wl2 = [load('data/wavelengths.npz')[k] for k in ['wl1','wl2']]
bias1 = load(join(datadir,'bias_y.npz'))['ccd1_denoised']
bias2 = load(join(datadir,'bias_y.npz'))['ccd2_denoised']
flat1 = load(join(datadir,'flat_ccd1.npz'))['flat'] - bias1[:,newaxis]

f = pf.open(join(datadir,'object','0000685605-20140707-OSIRIS-OsirisLongSlitSpectroscopy.fits'))
head = f[0].header
ccd1 = f[1].data - bias1[:,newaxis]
ccd2 = f[2].data - bias2[:,newaxis]
f.close()
ccds = concatenate([ccd1,ccd2], axis=1)
```

```python
pbls_f = r_[pb_full[:,0],pb_full[-1,-1]]
pbls_k = r_[pb_KI[:,0],pb_KI[-1,-1]]
```

```python
sky1 = ccd1[:,ssl1].mean(1)[31:-30]
sky2 = ccd2[:,ssl2].mean(1)[31:-30]

with sb.axes_style('ticks'):
    fig,ax = subplots(1,1,figsize=(AAOCW,0.6*AAOCW))
    [ax.axvline(pbl, lw=1, ls='-', alpha=0.1) for pbl in pbls_f]
    ax.plot(wl1, (ccd1[:,psl1].mean(1)[31:-30] - sky1)/4000., lw=1, c=c_bo)
    ax.plot(wl2, (ccd2[:,psl2].mean(1)[31:-30] - sky2)/4000., lw=1, c=c_ob)
    [ax.text(542.5+25*i, 0.04, i+1, ha='center') for i in range(16)]
sb.despine(fig)
setp(ax, xlabel='Wavelength [nm]', xlim=(500,950), ylabel='Normalised flux')
fig.tight_layout()
fig.savefig('plots/spectra.pdf')
```

## Light curves

```python
cols = ("JD flux target_flux comparison_flux RADEG DECDEG ROTANG AZIMUTH "
        "AIRMASS HUMIDITY PRESSURE TAMBIENT EXPTIME DATE-OBS UT_START filename").lower().split()

dt = loadtxt('data/tres_0003-gtc_osiris-wlc.dat', converters = {13:datestr2num, 14:lambda s:0, 15:lambda s:0})
df = pd.DataFrame(dt, columns=cols)
df['time'] = df.jd - int(df.jd[0])
```

```python
with sb.axes_style('ticks'):
    fig,ax = subplots(1,1, figsize=(AAOCW,0.6*AAOCW))
    ax.plot(df.time, df.target_flux/df.target_flux[0], c=cob, lw=1)
    ax.plot(df.time, df.comparison_flux/df.comparison_flux[0]+0.015, c=cbo, lw=1)
    setp(ax, xlim=(df.time.iloc[[0,-1]]), ylabel='Normalised flux', xlabel='JD - {:d}'.format(int(df.jd[0])))
sb.despine(fig)
fig.tight_layout()
fig.savefig('plots/raw_white_fluxes.pdf')
```

```python
fig,ax = subplots(1,1, figsize=(AAOCW,0.5*AAOCW))
plot(df.time, df.flux, c=cob)
setp(ax, xlim=(df.time.iloc[[0,-1]]), ylabel='Normalised flux', xlabel='JD - {:d}'.format(int(df.jd[0])))
fig.tight_layout()
fig.savefig('plots/white_relative_flux.pdf')
```
