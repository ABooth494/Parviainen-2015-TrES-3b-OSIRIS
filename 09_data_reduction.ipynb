```python
from glob import glob
import pyfits as pf
from scipy.ndimage import median_filter as mf
from george import GP
from george.kernels import ExpSquaredKernel
from os.path import join

datadir = '../../../data/GTC9-14A/OB0008/'
```

```python
biases = sorted(glob(join(datadir,'bias/*.fits')))[1:] 
flats = sorted(glob(join(datadir,'flat/*.fits')))
```

## Bias

### Notes

 - The first bias frame deviates strongly from the others, I'm leaving it out.
 - I'm creating two versions of bias vectors (in the dispersion direction)
   - Raw
   - Denoised using a GP

```python
def create_bias(bframes,mask):
    bs = bframes[:,:,mask].mean(1)  ## Calculate the mean along the dispersion axis
    median_b  = mf(bs, [30,1])
    abs_dev   = abs(bs - median_b)
    return where(abs_dev>0.5, median_b, bs).mean(1)
```

```python
b1 = dstack([pf.getdata(bias,1) for bias in biases])
b2 = dstack([pf.getdata(bias,2) for bias in biases])
```

```python
b1mask = ones(b1.shape[2], np.bool)
b2mask = ones(b2.shape[2], np.bool)
b2mask[20] = 0                        # Theres something funny in this frame, let's drop it
```

```python
b1_raw = create_bias(b1,b1mask)
b2_raw = create_bias(b2,b2mask)
```

```python
x = arange(b1_raw.size)
gp = GP(ExpSquaredKernel(500))
gp.compute(x, 1e-2)
```

```python
b1_denoised = gp.predict(b1_raw-b1_raw.mean(),x,mean_only=True)+b1_raw.mean()
b2_denoised = gp.predict(b2_raw-b2_raw.mean(),x,mean_only=True)+b2_raw.mean()
```

```python
savez(join(datadir,'bias_y'), ccd1_raw=b1_raw, ccd2_raw=b2_raw, 
      ccd1_denoised=b1_denoised, ccd2_denoised=b2_denoised)
```
